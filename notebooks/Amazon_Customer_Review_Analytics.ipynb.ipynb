## 📝 14.8 Lemmatization

Different grammatical forms of the same word often appear in customer reviews. For example, words such as **"works"**, **"worked"**, and **"working"** all represent the same underlying concept.

Lemmatization converts these words to their base (dictionary) form, known as the **lemma**. Unlike stemming, lemmatization considers the linguistic meaning of words, producing more accurate and readable results.

This standardization reduces vocabulary size while preserving the original meaning of customer reviews, making the text more suitable for Exploratory Text Analysis (ETA) and sentiment analysis.

---

### 🎯 Purpose

This step aims to:

- Convert words to their base forms.
- Reduce vocabulary redundancy.
- Preserve the linguistic meaning of words.
- Improve word frequency analysis.
- Produce standardized textual data for NLP.

---

```python
# ==========================================
# Apply Lemmatization
# ==========================================

lemmatizer = WordNetLemmatizer()

df_nlp["tokens"] = df_nlp["tokens"].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)

print("✅ Lemmatization completed successfully.")
```

```python
# ==========================================
# Verify Lemmatization
# ==========================================

df_nlp["tokens"].head(3)
```

### 🔍 Observation

The verification confirms that words have been successfully converted to their base forms using lemmatization.

For example, plural nouns have been converted to their singular forms, and many inflected words now share a common representation. This reduces vocabulary duplication while preserving the semantic meaning of customer reviews.

The resulting tokens are more consistent and better suited for word frequency analysis, keyword extraction, Word Cloud generation, and sentiment analysis.

### 📌 Key Findings

- Words have been converted to their base forms.

Unlike sentiment analysis, which focuses on determining whether customer opinions are positive, negative, or neutral, Exploratory Text Analysis aims to understand **what customers are discussing**, **how they express their opinions**, and **which topics dominate customer conversations**.

By analyzing review length, frequently occurring words, common phrases, and recurring discussion patterns, ETA transforms unstructured customer reviews into meaningful business intelligence.

The insights generated during this stage provide valuable context for understanding customer behavior and establish a strong foundation for the sentiment analysis performed later in this project.

---

By identifying the most frequently occurring words in customer reviews, we can understand the primary topics customers discuss, the product features they mention most often, and the recurring themes present in their feedback.

This analysis provides valuable business insights and forms the basis for more advanced text analyses such as Word Clouds, Bigrams, and Trigrams.

### 🎯 Business Question

### 🎯 Purpose

---

### 🔢 15.3.1 Calculate Word Frequency

The first step in word frequency analysis is to count how many times each word appears across all processed customer reviews.

Word frequency provides a quantitative overview of customer discussions and helps identify the most commonly mentioned product features, brands, and customer experiences.

By examining the most frequent words, we can gain an initial understanding of the dominant topics before performing more advanced phrase-level analyses.

---

```python
# ==========================================
# Calculate Word Frequency
# ==========================================

word_counts = Counter()

for tokens in df_nlp["tokens"]:
    word_counts.update(tokens)

print("✅ Word frequency calculated successfully.")
```

### ✅ Verification

Display the most frequently occurring words to verify that word frequencies have been calculated correctly.

```python
# ==========================================
# Verify Top 10 Most Frequent Words
# ==========================================

word_counts.most_common(10)
```

### 🔍 Initial Observation

The initial word frequency analysis successfully identifies the most frequently occurring words in the customer reviews.

Several product-related words such as **card**, **memory**, **phone**, and **sandisk** appear prominently, indicating that customers frequently discuss storage devices, compatibility, and product performance.

However, some high-frequency words such as **one** and **use** are generic in nature. Although they occur frequently, they provide limited business value because they do not describe specific product features or customer opinions.

These generic words can reduce the interpretability of the analysis by occupying positions that could otherwise highlight more informative keywords.

---

### 📌 Initial Findings

- Product-related words dominate customer discussions.
- Generic words such as **one** and **use** appear among the most frequent terms.
- The initial results provide a useful overview but can be refined further to improve interpretability.

---

### 🔧 Refining the Analysis

The initial frequency analysis revealed that a small number of frequently occurring words contribute little business insight.

Words such as **one**, **use**, **get**, **would**, **also**, **could**, and **even** are not traditional English stopwords, but they are overly generic within this dataset. They occur frequently across reviews without describing specific product characteristics, customer experiences, or opinions.

To improve the quality of the analysis, these dataset-specific words are treated as **custom stopwords** and removed before recalculating word frequencies.

This refinement allows more meaningful product-related terms to appear in the final results, providing clearer business insights.

---

```python
# ==========================================
# Remove Custom Stopwords
# ==========================================

custom_stopwords = {
    "one",
    "use",
    "get",
    "would",
    "also",
    "could",
    "even"
}

df_nlp["tokens"] = df_nlp["tokens"].apply(
    lambda words: [word for word in words if word not in custom_stopwords]
)

print("✅ Domain-specific stopwords removed successfully.")
```

```python
# ==========================================
# Recalculate Word Frequencies
# ==========================================

word_counts = Counter()

for tokens in df_nlp["tokens"]:
    word_counts.update(tokens)

print("✅ Word frequencies recalculated successfully.")
```

### ✅ Verification

Display the updated top ten most frequent words after removing the custom stopwords.

```python
# ==========================================
# Verify Updated Top 10 Most Frequent Words
# ==========================================

word_counts.most_common(10)
```

### 🔍 Updated Observation

After removing the custom stopwords, the updated word frequency analysis provides a clearer representation of customer discussions.

Generic words have been replaced by more informative terms such as **galaxy** and **speed**, which better reflect the products, features, and performance aspects discussed by customers.

The refined results improve the interpretability of the analysis and provide a stronger foundation for subsequent visualizations, phrase analysis, and business insight generation.

---

### 📌 Updated Key Findings

- Generic words have been successfully removed.
- Product-related keywords now dominate the frequency distribution.
- Storage devices, brand names, and performance features are the primary discussion topics.
- The refined vocabulary provides more meaningful business insights.
- The dataset is now ready for visualization using a bar chart.

---

> 💡 **Analyst Note:** Exploratory Text Analysis is an iterative process. Initial results often reveal generic but frequent words that are not useful for business interpretation. Refining the vocabulary by removing carefully selected custom stopwords improves the quality of insights while preserving meaningful customer opinions.

## 🔗 15.5 Most Frequent Bigrams

### 🔗 15.5.1 Generate Bigrams

A **bigram** is a sequence of two consecutive words that appear together in a sentence.

Unlike individual word analysis, bigrams preserve local context by capturing meaningful two-word phrases commonly used by customers. These phrases often represent product names, device compatibility, product features, or recurring customer experiences.

The generated bigrams will be used to calculate phrase frequencies and identify the most commonly occurring two-word combinations discussed in customer reviews.

---

```python
# ==========================================
# Import Bigram Utility
# ==========================================

from nltk.util import bigrams
```

```python
# ==========================================
# Generate Bigrams
# ==========================================

bigram_counts = Counter()

for tokens in df_nlp["tokens"]:
    bigram_counts.update(bigrams(tokens))

print("✅ Bigrams generated successfully.")
```

### ✅ Verification

Display the ten most frequently occurring bigrams to verify that two-word combinations have been generated successfully.

```python
# ==========================================
# Verify Top 10 Most Frequent Bigrams
# ==========================================

bigram_counts.most_common(10)
```

The generated bigrams reveal meaningful two-word phrases that frequently occur together within customer reviews.

Phrases such as **sd card**, **memory card**, **work great**, **samsung galaxy**, and **micro sd** appear prominently, indicating that customers naturally discuss products, compatibility, and performance using short descriptive phrases rather than isolated words.

Compared with individual word frequencies, bigrams provide richer contextual information and improve the interpretation of customer discussions.

---

### 📌 Key Findings

- Meaningful two-word phrases have been generated successfully.
- Product names and storage-related phrases dominate customer discussions.
- Contextual information is preserved more effectively than individual word analysis.
- The generated bigrams are ready for visualization.

> 💡 **Analyst Note:** Bigram analysis provides valuable contextual information because it identifies words that frequently occur together. This helps distinguish meaningful customer expressions from isolated high-frequency words.

These three-word combinations provide richer context than individual words or bigrams, making customer discussions easier to interpret.

### 🔗 15.6.1 Generate Trigrams

A **trigram** is a sequence of three consecutive words that appear together in a sentence.

Compared with bigrams, trigrams provide greater context by capturing longer expressions that customers naturally use when describing products, compatibility, performance, and their overall experiences.

The generated trigrams will be used to calculate phrase frequencies and identify the most frequently occurring three-word combinations in customer reviews.

---

```python
# ==========================================
# Import Trigram Utility
# ==========================================

from nltk.util import trigrams
```

```python
# ==========================================
# Generate Trigrams
# ==========================================

trigram_counts = Counter()

for tokens in df_nlp["tokens"]:
    trigram_counts.update(trigrams(tokens))

print("✅ Trigrams generated successfully.")
```

### ✅ Verification

The generated trigrams are stored in a frequency counter and will be used to identify the most frequently occurring three-word phrases across customer reviews.

```python
# ==========================================
# Verify Top 10 Most Frequent Trigrams
# ==========================================

trigram_counts.most_common(10)
```

### 🔍 Observation

Unlike bigrams, trigrams preserve additional contextual information by capturing longer customer expressions. These phrases often represent complete product names, device compatibility, and detailed descriptions of product features.

## 😊 16.2 Calculate Sentiment Polarity

Before classifying customer reviews into sentiment categories, each review is assigned a **sentiment polarity score**.

A polarity score is a numerical value that represents the emotional tone expressed in a piece of text. In this project, **TextBlob** is used to calculate the polarity of each cleaned customer review.

The polarity score ranges from **-1** to **+1**:

| Polarity Score | Interpretation |
|---------------:|----------------|
| **-1.0** | Extremely Negative |
| **0.0** | Neutral |
| **+1.0** | Extremely Positive |

Reviews with positive language receive higher polarity scores, while reviews containing negative opinions receive lower scores.

These polarity scores will later be used to classify customer reviews into **Positive**, **Neutral**, and **Negative** sentiment categories.

---

### 🎯 Business Question

### 🔹 16.2.1 Calculate Sentiment Polarity

The sentiment polarity of each cleaned customer review is calculated using **TextBlob**.

TextBlob analyzes the words contained within each review and assigns a polarity score between **−1** and **+1**, where negative values indicate negative sentiment, positive values indicate positive sentiment, and values close to zero indicate neutral sentiment.

The calculated polarity score is stored in a new column named **polarity** for subsequent analysis.

```python
# ==========================================
# Calculate Sentiment Polarity
# ==========================================

df_nlp["polarity"] = df_nlp["clean_review"].apply(
    lambda review: TextBlob(review).sentiment.polarity
)

print("✅ Sentiment polarity calculated successfully.")
```

## 🏷️ 16.3 Classify Customer Sentiment

While polarity scores provide a numerical measure of customer sentiment, they are not always intuitive for business stakeholders to interpret.

To simplify the analysis, each customer review is classified into one of three sentiment categories:

- **Positive**
- **Neutral**
- **Negative**

This transformation converts continuous polarity scores into discrete sentiment labels, making it easier to summarize customer opinions, visualize sentiment distribution, and compare textual sentiment with customer ratings.

---

### 🎯 Business Question

**How many customer reviews express positive, neutral, and negative sentiment?**

---

### 🎯 Purpose

This analysis aims to:

- Convert polarity scores into meaningful sentiment categories.

### 🏷️ 16.3.1 Define Sentiment Classification Rules

Before assigning sentiment labels, classification rules are established based on the polarity score generated by TextBlob.

The following thresholds are used:

| **Polarity Score** | **Sentiment** |
|-------------------:|---------------|
| Greater than **0.05** | Positive |
| Between **−0.05** and **0.05** | Neutral |
| Less than **−0.05** | Negative |

A small neutral interval around zero is intentionally used because polarity values very close to zero usually represent weak or balanced opinions rather than clearly positive or negative sentiments.

These thresholds help reduce unnecessary misclassification and produce more reliable sentiment categories.

---

```python
# ==========================================
# Create Sentiment Classification Function
# ==========================================

def classify_sentiment(polarity):

    if polarity > 0.05:
        return "Positive"

    elif polarity < -0.05:
        return "Negative"

    else:
        return "Neutral"
```

### 🏷️ 16.3.2 Assign Sentiment Labels

The sentiment classification function is now applied to every customer review.

Each polarity score is converted into one of the three sentiment categories, and the resulting label is stored in a new column named **sentiment**.

This categorical variable will be used throughout the remaining analyses to study customer satisfaction patterns and compare textual sentiment with customer ratings.

---

```python
# ==========================================
# Assign Sentiment Labels
# ==========================================

df_nlp["sentiment"] = df_nlp["polarity"].apply(classify_sentiment)

print("✅ Sentiment labels assigned successfully.")
```

## ⚠️ 21.1 Lexicon-Based Sentiment Analysis

This project utilized **TextBlob**, a lexicon-based sentiment analysis tool that determines sentiment based on predefined word polarity scores.

Although TextBlob performs well for general sentiment analysis, it may not always understand the complete context of customer reviews. Reviews containing sarcasm, negation, mixed opinions, or domain-specific expressions may occasionally be classified incorrectly.

### Example from This Project

During the analysis, a few **5-star customer reviews** were classified as **Negative** because they contained words such as **"bad"** or **"worst"**, even though the overall customer opinion was positive.

This demonstrates that lexicon-based approaches analyze individual word polarity more effectively than overall sentence meaning.

---

## ⚠️ 21.2 Single Product Dataset

The analysis was conducted using reviews from a single Amazon product.

Customer opinions, discussion topics, and sentiment patterns may differ across other product categories, brands, or industries.

Therefore, the findings should not be generalized to all Amazon products without additional analysis.

---

## ⚠️ 21.3 English Language Reviews Only

The dataset primarily consisted of English-language customer reviews.

Consequently, the analytical approach may not accurately represent customer opinions expressed in other languages or regional dialects.

Supporting multilingual sentiment analysis would require additional language-specific preprocessing and sentiment models.

More specialized language models could better capture these contextual meanings.
